```python
# popular website : www.renren.com www.douyu.com 
# unpopular website : www.tensorflow.org
def cdf(txtfile):
    x,y,dataset = [],[],[]
    file = open(txtfile, "r")
    element = file.readlines()
    num = len(element) - 1
    URL = element[0]
    for i in range(1, num + 1):
        dataset.append(int(element[i]))
    dataset.sort()
    point = 0
    for i in range(num): 
        if dataset[i] >=30:
            dataset[num-1] = 40
            break    
        if point != dataset[i]:
            x.append(point)
            y.append((float)(i)/num)
            point = dataset[i]
        i=i+1
    x.append(dataset[num-1])
    y.append(1)
    plt.plot(x, y, '', label = URL)


if __name__ == "__main__":
    import matplotlib.pyplot as plt
    plt.xlim(0, 35)
    plt.ylim(0, 1)
    plt.xlabel("times")
    plt.ylabel("probabilities")
    cdf("www.tensorflow.org.txt")
    cdf("www.douyu.com.txt")
    cdf("www.renren.com.txt")
    plt.legend(loc="upper left") 
    plt.savefig('hw2.png')
    plt.show()
```

from the picture above, we can observe that the gap times of tensorflow.org are relatively bigger than the other, this is make sense cause we can see that most people just download tensorflow for deep learning study but the proportion of people are small so there will be less people who will browsing tensorflow wesite that people visiting other's. As we can see the www.douyu.com are popular website so it just have a really small gap times
